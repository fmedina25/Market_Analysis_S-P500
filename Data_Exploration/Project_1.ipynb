```python
# Import the required libraries and dependencies
import alpaca_trade_api as tradeapi
from dotenv import load_dotenv
import os
import pandas as pd
```

```python
# Load the environment variables from the .env file 
# by calling the Load_dotenv function
load_dotenv()
```

```python
# Set the variables for the Alpaca API Key, Alpaca Secret Key, and the Alpaca Url address.
key_id = os.getenv('ALPACA_API_KEY')
secret_key = os.getenv('ALPACA_SECRET_KEY')
base_url = 'https://paper_api.alpaca.markets'
```

```python
# Create the tradeapi.Rest object
api = tradeapi.REST(
    key_id,
    secret_key,
    api_version="v2"
    )
```

```python
# Set The Tickers for AAPL, MSFT, AMZN, FB, and GOOGL
tickers = ["AAPL", "MSFT", "AMZN","FB", "GOOGL"]
```

```python
# Set your start and end dates as well as your isoformat
start_date_2018_01_01 = pd.Timestamp("2018-01-01", tz="America/New_York").isoformat()
end_date_2018_04_01 = pd.Timestamp("2018-04-01" , tz="America/New_York").isoformat()
```

```python
# Set timeframe to 1D
timeframe = "1D"
```

```python
#Use the current get_barset function to get current closing prices for the 5 stocks
def pull_data_1(x, y ):
    data_pull=api.get_barset(
            tickers,
            timeframe,
            start = x,
            end = y
        ).df
    return(data_pull)

```

```python
# Set the start and end date to pull your data for four months at a time
stock_data_2018_01_04 = pull_data_1(start_date_2018_01_01, end_date_2018_04_01)
```

```python
stock_data_frame_list = [
    stock_data_2018_01_04,
    stock_data_2018_04_07,
    stock_data_2018_07_11, 
    stock_data_2018_11_03, 
    stock_data_2019_03_07, 
    stock_data_2019_07_11,
    stock_data_2019_11_03,
    stock_data_2020_03_07,
    stock_data_2020_07_11,
    stock_data_2020_11_03,
    stock_data_2021_03_07,
    stock_data_2021_07_08
]
top_5_stocks = pd.concat(stock_data_frame_list)
```

```python
# Review the first five and last five rows
display(top_5_stocks.head())
display(top_5_stocks.tail())
```

```python
def pull():
    top_5_stocks_closing = pd.DataFrame()
    for ticker in tickers:
        top_5_stocks_closing[ticker] = top_5_stocks[ticker]["close"]
    columns_1 = ['AAPL_Close','MSFT_Close','AMZN_Close','FB_Close','GOOGL_Close']
    top_5_stocks_closing.columns = columns_1

    result = pd.concat([top_5_stocks_closing], axis=1)

    return(result)
```

```python
# Review the first five and last five rows
top_5_stocks_close = pull()
top_5_stocks_close.head()
```

```python
# Calculate the daily Returns
top_5_stocks_daily_returns = top_5_stocks_close.pct_change().dropna()
top_5_stocks_daily_returns.columns = tickers

top_5_stocks_daily_returns.head()
```
