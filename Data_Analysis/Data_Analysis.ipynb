```python
# Import the required libraries and dependencies
import pandas as pd
import hvplot.pandas
from pathlib import Path
import numpy as np
```

```python
Top_5_Stocks_df = pd.read_csv(
    Path('../Resources/Top_5_stocks_closing_price.csv')

)
```

```python
stocks_df = Top_5_Stocks_df.set_index('time')
stocks_df.head(
    
)
```

```python
#Calculate Daily Returns
stocks_daily_returns = stocks_df.pct_change() 
columns = ['AAPL_Returns','MSFT_Returns','AMZN_Returns','FB_Returns','GOOGL_Returns']
stocks_daily_returns.columns = columns
```

```python
# Review first five rows and drop NAN 
stocks_daily_returns.tail().dropna()
```

```python
stocks_daily_returns_2018 = stocks_daily_returns.loc['2018-01-03 00:00:00-05:00':'2018-12-31 00:00:00-05:00']
stocks_daily_returns_2019 = stocks_daily_returns.loc['2019-01-03 00:00:00-05:00':'2019-12-31 00:00:00-05:00']
stocks_daily_returns_2020_21 = stocks_daily_returns.loc['2020-01-02 00:00:00-05:00':]
```

```python
# Use the Pandas hvplot funtion to plot the daily return data for the 3-year time
# stocks_daily_returns.hvplot.box(
#     title = "Top 5 Stocks S&P 500 Daily Returns",
#     invert=True,
    
#     )
display(stocks_daily_returns_2018.hvplot.box(title = "Daily Returns for 2018", rot=45))
display(stocks_daily_returns_2019.hvplot.box(title = "Daily Returns for 2019", rot=45))
display(stocks_daily_returns_2020_21.hvplot.box(title = "Daily Returns for 2020", rot=45))
# stocks_daily_returns_2019.plot.box(ax=ax)
# stocks_daily_returns_2020_21.plot.box(ax=ax)
```

```python
# Calculate the cumulative returns using the 'cumprod' function
cumulative_returns = (1 + stocks_daily_returns).cumprod().dropna()
columns = ['AAPL_Cumulative','MSFT_Cumulative','AMZN_Cumulative','FB_Cumulative','GOOGL_Cumulative']
cumulative_returns.columns = columns
```

```python
cumulative_returns.describe()
```

```python
cumulative_returns.hvplot.line(
    height = 700,
    width = 1200, 
    title = "Cumulative Returns Top 5 Stocks S&P 500 During Covid From 2018-01 - 2021-08")
```
